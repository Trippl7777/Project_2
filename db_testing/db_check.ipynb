# Analysizing Crime Data

```python
# Dependencies and Setup
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
file = "crimemap.csv"
crime = pd.read_csv(file)
crime.head()
```

```python
file = "weathermap.csv"
weather = pd.read_csv(file)
weather.head()
```

```python
file = "population.csv"
population = pd.read_csv(file)
population.head()
```

```python
import sqlalchemy as db
```

```python
engine = db.create_engine('sqlite:///test/test.sqlite')
```

```python
crime.to_sql('crime', engine, if_exists='replace')
```

```python
weather.to_sql('weather', engine, if_exists='replace')
```

```python
population.to_sql('population', engine, if_exists='replace')
```

```python
sql_string = 'SELECT * FROM crime'
pd.read_sql_query(sql_string,engine)
```

```python
sql_string = 'SELECT * FROM weather'
pd.read_sql_query(sql_string,engine)
```

```python
sql_string = 'SELECT * FROM population'
pd.read_sql_query(sql_string,engine)
```
