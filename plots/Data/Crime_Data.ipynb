```python
# get file names from folder
file = "atlanta\\data_atlanta.csv"
atl_df = pd.read_csv(file)
atl_df["code"] = "A" + atl_df["code"].astype("str")
#atl_df.info()
atl_df.head()
```

```python
file = "boston\\data_boston.csv"
bos_df = pd.read_csv(file)
bos_df["code"] = "A" + bos_df["code"].astype("str")
#bos_df.info()
bos_df.head()
```

```python
file = "chicago\\data_chicago.csv"
chi_df = pd.read_csv(file)
chi_df["code"] = "A" + chi_df["code"].astype("str")
#chi_df.info()
chi_df.head()
```

```python
file = "denver\\data_denver.csv"
den_df = pd.read_csv(file)
den_df["code"] = "A" + den_df["code"].astype("str")
#bos_df.info()
den_df.head()
```

```python
file = "LA\\data_la.csv"
los_df = pd.read_csv(file)
los_df["code"] = "A" + los_df["code"].astype("str")
#los_df.info()
los_df.head()
```

```python
len(atl_df.index) + len(bos_df.index) + len(chi_df.index) + len(den_df.index) + len(los_df.index)
```

```python
df = pd.DataFrame()

df = df.append(atl_df,sort=False)
del atl_df
df = df.append(bos_df,sort=False)
del bos_df
df = df.append(den_df,sort=False)
del den_df
df = df.append(chi_df,sort=False)
del chi_df
df = df.append(los_df,sort=False)
del los_df

df.head()
```

```python
df.to_csv("combined_crime.csv", sep=',', encoding='utf-8')
```

```python
file = "crime_keys.csv"
keys_df = pd.read_csv(file)
keys_df["code"] = keys_df["code"].astype("object")
keys_df.head()
```

```python
crime = pd.merge(df, keys_df, how='inner', left_on = ["city", "code"], right_on = ["city", "code"])
crime.info()
crime.head()
```

```python
crime.to_csv("combined_crime_withKey.csv", sep=',', encoding='utf-8')
```

```python
file = "weather\\weather_df.csv"
weather_df = pd.read_csv(file)
weather_df = weather_df.drop_duplicates(subset=["city","date","time"],keep="first")
weather_df.info()
```

```python
crime_weather = pd.merge(crime, weather_df, how='left', left_on = ["city", "date", "time"], right_on = ["city", "date", "time"])
```

```python
crime_weather.info()
```

```python
cw_df = crime_weather.drop(crime_weather.columns[6], axis=1).copy()
cw_df.head()
```

```python
cw_df.columns = ["City", "Code", "StartDate", "StartTime", "Latitude", "Longitude", "Mapping", "Weather"]
cw_df.info()
```

```python
cw_df.to_csv("combined_crime_withKey_withweather.csv", sep=',', encoding='utf-8')
```
